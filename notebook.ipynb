```python
import pytesseract
from PIL import Image
import openai
import os
from pprint import pprint
import json
```

```python
json_file_path = 'api_key.json'
```

```python
with open(json_file_path, 'r') as json_file:
    data = json.load(json_file)
    api_key = data['api_key']
```

```python
text = pytesseract.image_to_string(Image.open('image1.png'))
```

```python
openai.api_key = api_key
```

```python
openai.Model.list()
```

```python
def get_prompt():
    p = """ 
        I'll give you a text. Based on the text, and the next prompt, you should try to only use that text and make answers for the next prompt provided. 
        Do not make up stuff on your own when the prompt provided is irrevelant to the text. 
        You can use your general knowledge to answer the prompt outside of the text given only if the text is very well known, reputed and popular text. Else, understand what's provided to you in the text and answer the prompt to your best understanding. 
        The prompt could involve asking questions based on the text provided or can ask you to summarize the text.

        Text: 
        ===================
        """
    
    return p
```

```python
def extract_text(text, task = 'summarize'):
    prompt = get_prompt() + '\n' + text  
    prompt = prompt + '\n =========== \n' + "PROMPT:" + task 
    
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "user", "content": prompt},
        ],
        temperature = 0.6
    )

    content = response.choices[0]['message']['content']
    return content
```

```python
res = extract_text(text)
res
```

```python
pprint(extract_text(text))
```
